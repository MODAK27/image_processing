```python

# Python program to Edge detection 
# using OpenCV in Python 
# using Sobel edge detection 
# and laplacian method 
import cv2
import numpy as np 

#Capture livestream video content from camera 0 
cap = cv2.VideoCapture(0) 

while(1): 

	# Take each frame 
	_, frame = cap.read()
    


    #Convert to HSV for simpler calculations 
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) 
	
	# Calcution of Sobelx 
	sobelx = cv2.Sobel(frame,cv2.CV_64F,1,0,ksize=5) 
	
	# Calculation of Sobely 
	sobely = cv2.Sobel(frame,cv2.CV_64F,0,1,ksize=5) 
	
	# Calculation of Laplacian 
	laplacian = cv2.Laplacian(frame,cv2.CV_64F) 
	
	cv2.imshow('sobelx',sobelx) 
	cv2.imshow('sobely',sobely) 
	cv2.imshow('laplacian',laplacian) 
	k = cv2.waitKey(5) & 0xFF
	if k == 27: 
		break

cv2.destroyAllWindows() 

#release the frame 
cap.release() 
```

```python
#Contours mapping
# import the required library 
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 
img_1 = cv2.imread('Downloads/scarlett-johansson-cover.jpg')
imgray = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY) 
# cv2.imshow('dog image', img_1)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
ret, thresh = cv2.threshold(imgray, 127, 255, 0)#threhold can be set on only gray image
im2, contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# cv2.drawContours(img_1, contours, -1, (0,255,0), 3)  #on original image drawing contours original gets modified
# cv2.drawContours(imgray, contours, -1, (0,255,0), 3)  #on gray image drawing contours original gets modified

cnt = contours[0]
#convex_hull
hull = cv2.convexHull(cnt)#So to get a convex hull as in above image, following is sufficient

rect = cv2.minAreaRect(cnt)
box = cv2.boxPoints(rect)
box = np.int0(box)
cv2.drawContours(img_1,[box],-1,(0,255,0),3)

x,y,w,h = cv2.boundingRect(cnt)
cv2.rectangle(img_1,(x,y),(x+w,y+h),(0,255,0),2)

(x,y),radius = cv2.minEnclosingCircle(cnt)
center = (int(x),int(y))
radius = int(radius)
cv2.circle(img_1,center,radius,(0,255,0),2)

cv2.imshow('original',img_1)
cv2.imshow('gary_image',imgray)
cv2.imshow('cat image', im2)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import cv2 as cv
import numpy as np
img = cv.imread('Downloads\scarlett-johansson-cover.jpg')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray,50,150,apertureSize = 3)
lines = cv.HoughLines(edges,1,np.pi/180,200)
for line in lines:
    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    cv.line(img,(x1,y1),(x2,y2),(0,0,255),2)
cv.imwrite('houghlines3.jpg',img)
```

```python

#hough transform for lines 

import cv2 as cv
import numpy as np
img = cv.imread('Downloads\scarlett-johansson-cover.jpg')
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
edges = cv.Canny(gray,50,150,apertureSize = 3)
lines = cv.HoughLinesP(edges,1,np.pi/180,100,minLineLength=100,maxLineGap=10)
for line in lines:
    x1,y1,x2,y2 = line[0]
    cv.line(img,(x1,y1),(x2,y2),(0,255,0),2)
cv.imwrite('houghlines5.jpg',img)
cv.imshow('pic',img)
k = cv.waitKey(0) & 0xff
cv.waitKey(0)
cv.destroyAllWindows
```

```python

#hough transform for circles 

import numpy as np
import cv2 as cv
img = cv.imread('Downloads\scarlett-johansson-cover.jpg',0)
img = cv.medianBlur(img,5)
cimg = cv.cvtColor(img,cv.COLOR_GRAY2BGR)
circles = cv.HoughCircles(img,cv.HOUGH_GRADIENT,1,20,
                            param1=50,param2=30,minRadius=0,maxRadius=0)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
    # draw the center of the circle
    cv.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
cv.imshow('detected circles',cimg)
cv.waitKey(0)
cv.destroyAllWindows()
```
